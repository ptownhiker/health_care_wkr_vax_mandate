```python
cases_states_filtered = community_cases.drop(columns = ['countyFIPS', 'StateFIPS','County Name','1/22/2020','1/23/2020', '1/24/2020', '1/25/2020', '1/26/2020'])
cases_states_daily_total = cases_states_filtered.groupby('State').sum()
new_cases_daily_total = cases_states_daily_total.diff(axis = 1)

display(new_cases_daily_total)
```

```python
new_cases_daily_total_modified = new_cases_daily_total.T.reset_index()

new_cases_daily_total_modified = new_cases_daily_total_modified.assign(Weeks = new_cases_daily_total_modified['index']).drop(columns = 'index')

new_cases_daily_total_modified['Weeks'] = new_cases_daily_total_modified['Weeks'].astype('datetime64[ns]')

new_cases_weekly_total_method_3 = new_cases_daily_total_modified.resample('W-Sun', on='Weeks').sum()
#for future referenc, I can also add  label='right', closed = 'right', but the code already defaults to this.
display(new_cases_weekly_total_method_3.head(5))
```

```python
deaths = pd.read_csv('covid_deaths_usafacts data correction.csv')
```

```python
deaths.head()
```

```python
deaths_states = deaths.drop(columns = ['countyFIPS', 
                       'StateFIPS','County Name'])
deaths_states_daily_total = deaths_states.groupby('State').sum()
new_deaths_daily_total = deaths_states_daily_total.diff(axis = 1)


new_deaths_daily_total.head()
```

```python
new_deaths_daily_total_modified = new_deaths_daily_total.T.reset_index()

new_deaths_daily_total_modified = new_deaths_daily_total_modified.assign(Weeks = new_deaths_daily_total_modified['index']).drop(columns = 'index')

new_deaths_daily_total_modified['Weeks'] = new_deaths_daily_total_modified['Weeks'].astype('datetime64[ns]')

new_deaths_weekly_total_method_3 = new_deaths_daily_total_modified.resample('W-Sun', label='left', closed = 'left', on='Weeks').sum()

display(new_deaths_weekly_total_method_3.tail(5))
```

```python
deaths2 = new_deaths_weekly_total_method_3.reset_index()

#df = df.reset_index()
```
